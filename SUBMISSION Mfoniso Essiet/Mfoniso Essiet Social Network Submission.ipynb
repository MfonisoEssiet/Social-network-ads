```python
# How many people made a purchase (Purchased=1) and how many did not (Purchase
purchase_counts = data['Purchased'].value_counts()
print('Purchase Counts:\n', purchase_counts)
```

```python
# What is the overall percentage of people who made a purchase?
percentage_purchase = (purchase_counts[1] / purchase_counts.sum()) * 100
print(f'Percentage of people who made a purchase: {percentage_purchase:.2f}%')
```

```python
# What is the average age and estimated salary of those who made a purchase compared to those who did not?
average_age_purchase = data[data['Purchased'] == 1]['Age'].mean()
average_salary_purchase = data[data['Purchased'] == 1]['EstimatedSalary'].mean()
average_age_no_purchase = data[data['Purchased'] == 0]['Age'].mean()
average_salary_no_purchase = data[data['Purchased'] == 0]['EstimatedSalary'].mean()

print('Average Age of Purchase:', average_age_purchase)
print('Average Estimated Salary of Purchase:', average_salary_purchase)
print('Average Age of No Purchase:', average_age_no_purchase)
print('Average Estimated Salary of No Purchase:', average_salary_no_purchase)
```

```python
# Are there any missing values in the dataset, and if so, in which columns?
missing_values = data.isnull().sum()
print('Missing Values:\n', missing_values)
```

```python
# Exploratory Questions:
# Is there a correlation between age and the likelihood of making a purchase?
sns.boxplot(x='Purchased', y='Age', data=data)
plt.title('Age vs. Purchase')
plt.show()
```

```python
# Is there a correlation between estimated salary and the likelihood of making
sns.boxplot(x='Purchased', y='EstimatedSalary', data=data)
plt.title('Estimated Salary vs. Purchase')
plt.show()
```

```python
# Are there any outliers in the dataset, and how do they affect the analysis?
sns.boxplot(data=data[['Age', 'EstimatedSalary']])
plt.title('Boxplot of Age and Estimated Salary')
plt.show()
```

```python
# How does the distribution of estimated salary differ between those who made a purchase and those who did not?
sns.kdeplot(data[data['Purchased'] == 1]['EstimatedSalary'], label='Purchased')
sns.kdeplot(data[data['Purchased'] == 0]['EstimatedSalary'], label='Not Purchased')
plt.title('Estimated Salary Distribution')
plt.xlabel('Estimated Salary')
plt.ylabel('Density')
plt.legend()
plt.show()

```
